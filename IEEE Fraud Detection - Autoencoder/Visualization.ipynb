```
sns.set_palette(sns.color_palette("Spectral"))
```

```
plt.rcParams["figure.figsize"]=6,5
plt.title('Percent of Fraud',fontweight = 'bold')
sns.countplot(x = df['isFraud'] )
plt.show()
```

```
plt.rcParams["figure.figsize"]=13,5 
fig, ax =plt.subplots(1,2)

sns.countplot(data = df[df['isFraud'] == True], x = 'ProductCD' , ax = ax[0] ).set_title('Fraud',fontweight = 'bold')
sns.countplot(data = df[df['isFraud'] == False], x = 'ProductCD', ax = ax[1]).set_title('Not Fraud',fontweight = 'bold')

fig.show()
```

```
cols = ['card4','card6']

plt.rcParams["figure.figsize"]=13,10
fig, ax =plt.subplots(2,2)

for i in range(len(cols)):
  sns.countplot(data = df[df['isFraud']==True], x = cols[i],ax = ax[i,0]).set_title(cols[i]+' Fraud',fontweight = 'bold')

for i in range(len(cols)):
  sns.countplot(data = df[df['isFraud']==False], x = cols[i],ax = ax[i,1]).set_title(cols[i]+' Not Fraud',fontweight = 'bold')
  
fig.show()
```

```
cols = ['card1','card2','card3','card5']

plt.rcParams["figure.figsize"]=12,25
fig, ax =plt.subplots(4,2)

for i in range(len(cols)):
  sns.histplot(data = df[df['isFraud']==True], x = cols[i],ax = ax[i,0],shrink = 0.9,bins = 10).set_title(cols[i]+' Fraud',fontweight = 'bold')


for i in range(len(cols)):
  sns.histplot(data = df[df['isFraud']==False], x = cols[i],ax = ax[i,1],shrink = 0.9,bins = 10).set_title(cols[i]+' Not Fraud',fontweight = 'bold')

fig.show()
```

```
plt.rcParams["figure.figsize"]=23,16
fig, ax =plt.subplots(nrows = 2, ncols=1)

a = df['DeviceInfo'][df['isFraud'] == False].value_counts(sort = True)[:20]
tags = a.index

sns.barplot(x = tags , y = a, ax = ax[0], palette = 'rocket_r').set_title('Fraud',fontweight = 'bold')

a2 = df['DeviceInfo'][df['isFraud'] == True].value_counts(sort = True)[:20]
tags2 = a.index

sns.barplot(x = tags2 , y = a2 , ax = ax[1] , palette = 'rocket_r').set_title('Not Fraud',fontweight = 'bold')

for a in ax:
    matplotlib.pyplot.sca(a)
    plt.xticks(rotation=20)

fig.show()
```
